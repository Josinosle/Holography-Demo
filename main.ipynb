```python
import matplotlib.pyplot as plt
import numpy as np
import mediumpy as mp

ax = plt.figure().add_subplot(projection='3d')

medium = mp.Medium(10,10,10,1,1)
medium.draw(ax)


# setting up a parametric curve
t = np.arange(0, 2*np.pi+.1, 0.01)
x, y, z = 3*t, 5*t, 3*t


ax.plot3D(x,y,z)

plt.show()
```
